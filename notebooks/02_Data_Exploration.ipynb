```python
# Zelle 1 - Pfad definieren (einmal ändern, überall gültig)
from pathlib import Path

MATCH_ID = "1886347"
BASE_PATH = Path(f"/content/drive/MyDrive/SkillCorner_Dataset-ML_Reaerch/data/matches/{MATCH_ID}")

phases_path   = BASE_PATH / f"{MATCH_ID}_phases_of_play.csv"
dynamic_path  = BASE_PATH / f"{MATCH_ID}_dynamic_events.csv"
tracking_path = BASE_PATH / f"{MATCH_ID}_tracking_extrapolated.jsonl"
```

```python
# Zelle 2 - phases_of_play laden und verstehen
import pandas as pd

df_phases = pd.read_csv(phases_path)
df_phases.info(show_counts=True, verbose=True)
```

```python
# Zelle 4 - dynamic_events laden und verstehen
df_dynamic = pd.read_csv(dynamic_path)
df_dynamic.info(show_counts=True, verbose=True)
```

```python
# Zelle 6 - Tracking: erste Zeile anschauen
import json

with open(tracking_path) as f:
    erste_zeile = json.loads(f.readline())

print(f"Anzahl Schlüssel: {len(erste_zeile.keys())}")
print(erste_zeile.keys())
```

```python
# Zelle 7 - Tracking: Struktur jedes Schlüssels verstehen
for key, value in erste_zeile.items():
    print(f"{key}: {type(value).__name__} → {value}")
    print("---")
```

```python
# Zelle 8 - Tracking: player_data eines aktiven Frames anschauen
with open(tracking_path) as f:
    for i, line in enumerate(f):
        if i == 200:
            frame_200 = json.loads(line)
            break

print(f"Anzahl Spieler in Frame 200: {len(frame_200['player_data'])}")
print(json.dumps(frame_200["player_data"][0], indent=2))
```
